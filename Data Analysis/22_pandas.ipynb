# Methods

```python
df = pd.read_csv("tips.csv")
```

```python
def last_four(num):
  return str(num)[-4:]
df["last_found"] = df["CC Number"].apply(last_four)
```

```python
def dollar_sign(num):
  return "$" * int(np.floor(np.log10(num))+1)
df["total_bill"].apply(dollar_sign)
```

```python
def quality(total_bill, tip):
  if tip/total_bill > 0.25:
    return "generous"
  else:
    return "Other"
```

```python
df["tip_quality"] = df[["total_bill", "tip"]].apply(lambda df:quality(df["total_bill"], df["tip"]), axis=1)
```

```python
df[df["tip_quality"] == "generous"]
```

```python
df["tip_quality"] = np.vectorize(quality)(df["total_bill"], df["tip"])
```

```python
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
pd.merge(registrations, logins, how="inner", on="name")
```

```python
pd.merge(left=registrations, right=logins, how="left")
```

```python
pd.merge(registrations, logins, how="outer", on="name")
```

```python
registrations = registrations.set_index("name")
```

```python
pd.merge(registrations, logins, left_index=True,right_on="name")
```

```python
registrations = registrations.reset_index()
```

```python
registrations.columns = ["reg_name","reg_id"]
```

```python
pd.merge(registrations, logins, left_on="reg_name", right_on="name")
```
