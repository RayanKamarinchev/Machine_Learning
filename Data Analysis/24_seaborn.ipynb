```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("dm_office_sales.csv")
df
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="salary", y="sales", data= df,hue="level of education", size = "work experience", alpha=0.7, style = "level of education")
```

```python
sns.rugplot(x="salary", data=df, height=0.5)
```

```python
sns.displot(data=df, x="salary", bins=17, kde=True)
```

```python
sns.kdeplot(data=df, x="salary", clip=(40000, 160000), shade=True)
```

```python
plt.figure(figsize=(10,6))
sns.countplot(data=df, x="division", hue="level of education")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data = df, x = "level of education", y = "salary", estimator = np.mean)
```

```python
df = pd.read_csv("StudentsPerformance.csv")
df
```

```python
sns.jointplot(x="math score", y="reading score", data = df, kind="hist") #hex, hist, kde shade=true
```

```python
sns.jointplot(x="math score", y="reading score", data = df, kind="kde", shade=True)
```

```python
sns.jointplot(x="math score", y="reading score", data = df, hue="gender") 
```

```python
sns.pairplot(data=df, hue="gender", corner=True)
```

```python
sns.catplot(x="gender", y="math score", data=df, row="lunch")
```

```python
g = sns.PairGrid(df)
g = g.map_upper(sns.scatterplot)
g=g.map_lower(sns.kdeplot)
```

```python
sns.heatmap(df.drop("Life expectancy", axis=1), lw=2, annot=True, cmap="viridis")
```
