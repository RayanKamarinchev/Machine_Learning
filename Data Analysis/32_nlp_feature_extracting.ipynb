```python
with open("One.txt") as mytext:
  words = mytext.read().lower().split()
  uni_words = set(words)
with open("Two.txt") as mytext:
  words_2 = mytext.read().lower().split()
  uni_words_2 = set(words_2)
```

```python
all_uni_words = set()
all_uni_words.update(uni_words)
all_uni_words.update(uni_words_2)
```

```python
vocab = dict()
i=0
for word in all_uni_words:
  vocab[word] = i
  i+=1
```

```python
one_freq=[0] * len(vocab)
two_freq=[0] * len(vocab)
all_words = [''] * len(vocab)
```
